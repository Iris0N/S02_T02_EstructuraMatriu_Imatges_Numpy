## Exercici 5

```python
import numpy as np
arr = np.array([[3, 2, 3, 5, 6], [4, 6, 6, 7, 7]])
sumColumn = arr[:, 1].sum()
sumRow = arr[1, :].sum()
print("2nd column", arr[:,1], "total sum", 
      sumColumn,", 2nd row", arr[1, :], "total sum", sumRow)
```
